# SEMANA 6- CLustering

#**Definicion**

En diferentes contextos, el término "cluster" puede tener significados ligeramente diferentes:

**Informática y Tecnología:**
En el campo de la informática, un cluster es un conjunto de computadoras interconectadas que trabajan juntas como si fueran una sola entidad más poderosa. Los clusters informáticos se utilizan para realizar tareas que requieren un alto nivel de procesamiento, como cálculos científicos, simulaciones, análisis de datos masivos, etc. Estas computadoras pueden estar físicamente cercanas o distribuidas en diferentes ubicaciones geográficas.

**Estadística:**
En estadística, un cluster se refiere a grupos de observaciones o datos que son similares entre sí dentro del mismo grupo, pero diferentes de los datos en otros grupos. El análisis de clúster es una técnica estadística que se utiliza para agrupar datos en función de su similitud en características específicas. Esto puede ser útil para la segmentación de mercado, la agrupación de clientes en grupos con intereses similares, la clasificación de datos en conjuntos distintos, entre otros.

**Biología:**
En biología, un cluster puede referirse a un grupo de genes, proteínas o individuos que comparten características similares. El análisis de clúster en este contexto puede ayudar a identificar patrones genéticos, relaciones evolutivas o funciones biológicas comunes.

**Economía y Ciencias Sociales:**
En ciencias sociales y economía, el término "cluster" se utiliza para describir la concentración geográfica de empresas e industrias relacionadas. Por ejemplo, una zona en la que varias empresas tecnológicas se establecen en estrecha proximidad se considera un "cluster tecnológico", lo que puede promover la innovación y la colaboración.

#**En que se Basa el Clustering**

El proceso de clustering (agrupamiento) en estadística e informática se basa en la idea fundamental de agrupar objetos similares en conjuntos, donde los objetos dentro de cada conjunto son más similares entre sí que con los objetos en otros conjuntos. Este proceso es una técnica importante tanto en estadística como en informática para encontrar patrones y estructuras en datos que pueden no ser evidentes a simple vista.

La base del proceso de clustering implica encontrar una forma de medir la similitud entre los objetos, que pueden ser puntos de datos, elementos, observaciones, etc. La elección de la métrica de similitud o distancia es crucial y puede variar según el dominio de los datos y el objetivo del análisis.

#**Proceso del clustering**

**Definir el problema y los datos:**
Esto implica comprender el conjunto de datos, las características de los objetos y el objetivo del análisis de clustering. ¿Qué estás tratando de lograr al agrupar los datos? ¿Qué características son relevantes para la agrupación?

**Preparación de datos:**
En muchos casos, es importante preprocesar los datos para que sean comparables y significativos. Esto puede involucrar la normalización, la eliminación de características irrelevantes o la transformación de los datos en una forma más adecuada.

**Selección de características y métricas de distancia:**
Elegir las características relevantes para el clustering es importante. Además, seleccionar una métrica de distancia adecuada para medir la similitud entre los objetos es crucial. Esto podría ser la distancia euclidiana, la similitud de coseno, la distancia de Manhattan, entre otras.

**Elegir el algoritmo de clustering:**
Hay varios algoritmos de clustering disponibles, cada uno con sus propias ventajas y desventajas. Algunos de los algoritmos más comunes incluyen K-means, DBSCAN, Agglomerative Hierarchical Clustering, Mean Shift, etc.

**Elegir el número de clústeres:**
Determinar el número óptimo de clústeres puede ser un desafío. Esto puede hacerse utilizando métodos como el método del codo, el índice de Silhouette, o incluso conocimiento previo del dominio.

**Aplicar el algoritmo de clustering:**
Utilizar el algoritmo elegido para agrupar los datos en conjuntos basados en las características y la métrica de distancia definidas. Cada algoritmo puede tener parámetros que deben ajustarse para obtener resultados óptimos.

**Evaluar los resultados:**
Evaluar la calidad de los clústeres formados puede hacerse utilizando métricas como el índice de Silhouette, la cohesión y la separación de los clústeres, y la interpretación visual de los resultados.

**Interpretación y análisis de resultados:**
Una vez que los clústeres se han formado, es importante interpretar su significado. ¿Qué implicaciones tienen estos grupos? ¿Qué patrones o características comunes se encuentran en cada grupo?

#**Cluster de FRUTAS PLANO en 2D**

```python
####################################################
#9- CLUSTER:

############################
#Saber los criterios


############################
#Elegir el ALGORITMO DE CLUSTER a mis DATOS

#ELEGIREMOS K-MEANS

#######################################################################
# Importamos las herramientas que vamos a usar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans  # Esto es para hacer los grupos de frutas
from sklearn.preprocessing import StandardScaler  # Esto es para poner los datos en un formato especial
from sklearn.metrics import silhouette_score  # Esto nos ayuda a saber qué tan buenos son nuestros grupos



#########################################
#2- Crear o extraer los datos un de un data set:

# Imagina que tenemos información sobre diferentes frutas, como su tamaño, peso, dulzura y color.
# Aquí guardamos esos datos en una "tabla" llamada DataFrame.
data = np.array([[10, 300, 1.5, 0.8],   # Tamaño, peso, dulzura, color
                 [8, 250, 1.2, 0.7],    # Tamaño, peso, dulzura, color
                 [6, 200, 1.0, 0.6],    # Tamaño, peso, dulzura, color
                 [12, 350, 1.8, 0.9],   # Tamaño, peso, dulzura, color
                 [9, 280, 1.3, 0.75]])  # Tamaño, peso, dulzura, color

#########################################
#3- Crear los nombres para las columnas
feature_names = ['Tamaño', 'Peso', 'Dulzura', 'Color']


#########################################
#4- UNIR: Los nombres que lleva cada columnas y LA MATRIX - CREAR EL DATA FRAME
fruits_df = pd.DataFrame(data, columns=feature_names)  # Creamos una tabla para las frutas


#########################################
#5- PREPARAR LOS DATOS - Escala los distintos valores para que PUEDAN COMPARARSE:
    #Ej: 0.6 y 66 (int, float)

# Ahora vamos a preparar nuestros datos para que podamos trabajar con ellos.
# Imagina que ponemos los datos en una caja especial para que podamos medirlos mejor.
scaler = StandardScaler()  # Creamos una "caja" especial para poner los datos
fruits_scaled = scaler.fit_transform(fruits_df)  # Ponemos las frutas en la "caja"


#########################################
#6- SEPARAR DATOS EN GRUPOS segun sus CARACTERISTICAS
    #Es decir AILSA Los datos por caracteristicas

# Imagina que queremos separar las frutas en grupos según sus características.
# Decidimos que queremos 2 grupos. Esto es como si clasificáramos frutas por sus parecidos.
num_clusters = 2


#########################################
#7- Aplicamos LA MAQUINA = Aplicamos la FUNCION

# Usamos una "máquina" llamada K-Means para hacer los grupos de frutas. MAQUINA=FUNCION (Esta relacionado con el algoritmo que elijamos)
kmeans = KMeans(n_clusters=num_clusters)  # Creamos la "máquina" para hacer grupos
fruits_df['grupo'] = kmeans.fit_predict(fruits_scaled)  # Usamos la "máquina" para hacer los grupos


#########################################

# Ahora vamos a ver cómo quedaron nuestros grupos en un dibujo.
# Imagina que dibujamos las frutas en un papel y les ponemos colores para cada grupo.
plt.figure(figsize=(8, 4))
for grupo_id in range(num_clusters):
    grupo_data = fruits_df[fruits_df['grupo'] == grupo_id]
    plt.scatter(grupo_data['Tamaño'], grupo_data['Peso'], label=f'Grupo {grupo_id}')

plt.xlabel('Tamaño')
plt.ylabel('Peso')
plt.title('Grupos de Frutas')
plt.legend()
plt.show()


# Queremos saber si nuestros grupos son buenos. Usamos una puntuación especial llamada "Silhouette Score".
# Imagina que es como ponerle una nota a nuestros grupos.
silhouette_avg = silhouette_score(fruits_scaled, fruits_df['grupo'])  # Calculamos la nota

print(f"Nota de Silhouette: {silhouette_avg:.4f}")









```
